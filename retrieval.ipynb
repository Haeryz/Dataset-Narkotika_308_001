## Data Preprocessing

```python
# Step 1: Keep all data including empty cells (will be filled manually later)
df_clean = df_filtered.copy()

print("=== DATA PRESERVED ===")
print(f"Total rows: {len(df_clean)}")
print(f"Empty cells will be filled manually later")

display(df_clean.head())
```

```python
# Step 2: Convert non-null text columns to string and strip whitespace
text_columns = ['nomor', 'nama pengadilan negeri', 'petunjuk bb', 'putusan']

for col in text_columns:
    # Only process non-null values, keep NaN as is
    df_clean[col] = df_clean[col].apply(lambda x: str(x).strip() if pd.notna(x) else x)

print("=== TEXT COLUMNS CLEANED ===")
print("Non-null values converted to string and whitespace removed")
print("Empty cells preserved for manual filling later")
display(df_clean.head())
```

```python
# Step 3: Create a combined text field for vector indexing (only for rows with all data)
# For rows with missing data, combined_text will be empty/partial
def create_combined_text(row):
    parts = []
    
    if pd.notna(row['nomor']):
        parts.append(f"Nomor Putusan: {row['nomor']}")
    
    if pd.notna(row['nama pengadilan negeri']):
        parts.append(f"Pengadilan: {row['nama pengadilan negeri']}")
    
    if pd.notna(row['petunjuk bb']):
        parts.append(f"Barang Bukti: {row['petunjuk bb']}")
    
    if pd.notna(row['putusan']):
        parts.append(f"Putusan: {row['putusan']}")
    
    return "\n\n".join(parts) if parts else ""

df_clean['combined_text'] = df_clean.apply(create_combined_text, axis=1)

print("=== COMBINED TEXT FIELD CREATED ===")
print("Rows with missing data will have partial/empty combined_text")
print("\nSample combined text (first 500 chars) from a complete row:")
print("-" * 80)
# Find first row with complete data
complete_row = df_clean[df_clean['combined_text'].str.len() > 0].iloc[0] if len(df_clean[df_clean['combined_text'].str.len() > 0]) > 0 else None
if complete_row is not None:
    print(complete_row['combined_text'][:500] + "...")
print("\n" + "="*80)
```

```python
# Step 4: Reset index for clean indexing
df_clean = df_clean.reset_index(drop=True)

print("=== FINAL PREPROCESSED DATASET ===")
print(f"Total rows: {len(df_clean)}")
print(f"Total columns: {len(df_clean.columns)}")
print(f"\nColumns: {list(df_clean.columns)}")

display(df_clean.head())
```

## Inspect and Clean Petunjuk BB Column

```python
# Inspect the actual structure of petunjuk bb column
print("=== INSPECTING PETUNJUK BB STRUCTURE ===\n")

# Show raw values
for i in range(min(5, len(df_clean))):
    if pd.notna(df_clean.loc[i, 'petunjuk bb']):
        print(f"Row {i}:")
        print(f"Type: {type(df_clean.loc[i, 'petunjuk bb'])}")
        print(f"Value: {df_clean.loc[i, 'petunjuk bb']}")
        print("\n" + "-"*80 + "\n")
```

```python
# Parse the JSON array in petunjuk bb column
import json

def parse_petunjuk_bb(value):
    """Parse the JSON array string in petunjuk bb and convert to clean text"""
    if pd.isna(value):
        return value
    
    try:
        # Parse the JSON array
        items = json.loads(value)
        
        # Join items with numbered list
        if isinstance(items, list):
            numbered_items = [f"{i+1}. {item}" for i, item in enumerate(items)]
            return "\n".join(numbered_items)
        else:
            return str(value)
    except:
        # If parsing fails, return as is
        return str(value)

# Apply parsing to petunjuk bb
df_clean['petunjuk bb'] = df_clean['petunjuk bb'].apply(parse_petunjuk_bb)

print("=== PETUNJUK BB PARSED ===")
print("JSON arrays converted to numbered list format\n")
print("Sample from first row:")
print("-" * 80)
print(df_clean.loc[0, 'petunjuk bb'])
```

```python
# Recreate combined_text with properly parsed petunjuk bb
def create_combined_text_v2(row):
    parts = []
    
    if pd.notna(row['nomor']):
        parts.append(f"Nomor Putusan: {row['nomor']}")
    
    if pd.notna(row['nama pengadilan negeri']):
        parts.append(f"Pengadilan: {row['nama pengadilan negeri']}")
    
    if pd.notna(row['petunjuk bb']):
        parts.append(f"Barang Bukti:\n{row['petunjuk bb']}")
    
    if pd.notna(row['putusan']):
        parts.append(f"Putusan: {row['putusan']}")
    
    return "\n\n".join(parts) if parts else ""

df_clean['combined_text'] = df_clean.apply(create_combined_text_v2, axis=1)

print("=== COMBINED TEXT UPDATED ===")
print("Combined text recreated with properly parsed petunjuk bb\n")
print("Sample (first 600 chars):")
print("-" * 80)
if len(df_clean[df_clean['combined_text'].str.len() > 0]) > 0:
    print(df_clean[df_clean['combined_text'].str.len() > 0].iloc[0]['combined_text'][:600] + "...")
```

## Vector Store Setup

```python
from sentence_transformers import SentenceTransformer
embedding_model_name = "sentence-transformers/all-mpnet-base-v2"
embedding_model = SentenceTransformer(embedding_model_name)
vector_dim = embedding_model.get_sentence_embedding_dimension()
print(f"Loaded {embedding_model_name} with dimension {vector_dim}")
```

```python
# Prepare records with non-empty combined text
vector_texts = index_df["combined_text"].tolist()
vector_payloads = index_df.to_dict("records")
vector_ids = index_df["No"].astype(int).tolist()
print(f"Prepared {len(vector_texts)} records for embedding")
```

```python
# Create sentence embeddings
embeddings = embedding_model.encode(
    vector_texts,
    batch_size=32,
    show_progress_bar=True,
    convert_to_numpy=True,
    normalize_embeddings=True
)
print(f"Embedding matrix shape: {embeddings.shape}")
```

```python
points = []
for point_id, embedding, payload in zip(vector_ids, embeddings, vector_payloads):
    points.append(
        rest.PointStruct(id=point_id, vector=embedding.tolist(), payload=payload)
    )

batch_size = 64
total_points = len(points)
print(f"Uploading {total_points} points to {collection_name} in batches of {batch_size}")

for batch_start in range(0, total_points, batch_size):
    batch = points[batch_start:batch_start + batch_size]
    qdrant_client.upsert(
        collection_name=collection_name,
        points=batch,
        wait=True,
    )
    completed = min(batch_start + batch_size, total_points)
    print(f"Uploaded {completed}/{total_points} points")

print(f"Finished uploading {total_points} points into {collection_name}")
```

## Retrieval Quality Evaluation

```python
# Encode test queries and collect ranked results
k_values = [1, 3, 5, 10]
max_k = max(k_values)
test_queries = test_df["combined_text"].tolist()
test_embeddings = embedding_model.encode(
    test_queries,
    batch_size=16,
    show_progress_bar=True,
    convert_to_numpy=True,
    normalize_embeddings=True
)
metrics_bucket = {k: {"precision": [], "recall": [], "ndcg": []} for k in k_values}
hit_overview = []
for row, query_vector in zip(test_df.itertuples(index=False), test_embeddings):
    search_hits = qdrant_client.search(
        collection_name=collection_name,
        query_vector=query_vector.tolist(),
        limit=max_k,
        with_payload=False,
    )
    retrieved_ids = [int(point.id) for point in search_hits]
    retrieved_scores = [point.score for point in search_hits]
    relevances = [1 if rid == row.No else 0 for rid in retrieved_ids]
    hit_rank = relevances.index(1) + 1 if 1 in relevances else None
    for k in k_values:
        top_rel = relevances[:k]
        hit = 1 if any(top_rel) else 0
        metrics_bucket[k]["precision"].append(sum(top_rel) / k)
        metrics_bucket[k]["recall"].append(float(hit))
        if any(relevances):
            metrics_bucket[k]["ndcg"].append(ndcg_score([relevances], [retrieved_scores], k=k))
        else:
            metrics_bucket[k]["ndcg"].append(0.0)
    hit_overview.append({"query_id": row.No, "hit_rank": hit_rank})
hit_summary_df = pd.DataFrame(hit_overview)
display(hit_summary_df.head())
```

```python
# Summarise Precision@K, Recall@K, NDCG@K
metrics_summary = []
for k in k_values:
    metrics_summary.append({
        "K": k,
        "Precision@K": np.mean(metrics_bucket[k]["precision"]) if metrics_bucket[k]["precision"] else np.nan,
        "Recall@K": np.mean(metrics_bucket[k]["recall"]) if metrics_bucket[k]["recall"] else np.nan,
        "NDCG@K": np.mean(metrics_bucket[k]["ndcg"]) if metrics_bucket[k]["ndcg"] else np.nan,
    })
metrics_df = pd.DataFrame(metrics_summary)
display(metrics_df)
```

```python
# Plot ranking quality metrics across K values
plt.figure(figsize=(8, 4))
plt.plot(metrics_df["K"], metrics_df["Precision@K"], marker="o", label="Precision@K")
plt.plot(metrics_df["K"], metrics_df["Recall@K"], marker="s", label="Recall@K")
plt.plot(metrics_df["K"], metrics_df["NDCG@K"], marker="^", label="NDCG@K")
plt.title("Retrieval performance on held-out queries")
plt.xlabel("Top-K")
plt.ylabel("Score")
plt.ylim(0, 1.05)
plt.xticks(metrics_df["K"])
plt.grid(True, linestyle="--", alpha=0.4)
plt.legend()
plt.tight_layout()
plt.show()
```
